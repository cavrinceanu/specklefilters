```python
from osgeo import gdal
import numpy as np
import matplotlib.pyplot as plt
from skimage import exposure
import math
import cv2 as cv
```

```python
file="/home/cristina/Seeps/test/subset1.tif"
sar=gdal.Open(file)
band=sar.GetRasterBand(1)
band.GetMetadata()
arr=band.ReadAsArray()
[cols, rows]=arr.shape
arr=arr.astype(dtype='uint8')
```

```python
eq=exposure.equalize_hist(arr).astype(np.float32)
```

```python
despeckle = cv.bilateralFilter(eq,7,80,80)
```

```python
plt.figure(figsize=(15,15))
plt.imshow(despeckle, cmap='gray')
plt.axis('off')
```

```python
roi = eq[2000:2500, 1900:3200]
```

```python
plt.figure(figsize=(10,10))
plt.imshow(roi, cmap='gray')
plt.axis('off')
```
